```python
# PERCEPTRON PERMITE APENAS CALCULOS DISCRETOS de classificao 1 ou 0 (funo limiar)

import numpy as np
from matplotlib import pyplot as plt

X = np.array([[0,0],[0,1],[1,0],[1,1]])
#Y = np.array( [0 , 1 , 1, 1]) # OR lgico
Y = np.array([0,0,0,1]) # AND lgico
#Y = np.array([0,1,1,0])# XOR lgico, como o Perceptron apenas separa as classes com uma linha reta as classes 1 e 0, e no caso do XOR temos classes que NO so LNEARMENTE SEPARVEIS (sepatrradas por uma reta)

#X = (X.dot(X.T)+1)**2  # kernel


threshold = 0.6 # limiar

n = 0.0001 # taxa de aprendizado

W = np.random.rand(np.shape(X)[1]) # INCIALIZAO ALEATRIA PARA MOSTRAR COMO A INCIIALIZAO AO REDOR DOS CIRUCLOS DE PESO  PODE SER DETERMINANE PARA O TEMPO DE ATUALIZADO CODIGO
flag = 1
cont=0
while(flag==1):
    flag=0
    print('#iteration %d'%cont)
    for i in range(len(X)):
        net = X[i].dot(W)  # Somatorio peso x entrada

        if net>=threshold: # FUNO LIMIAR, SE A SOMATORIO DOS PESOS*ENTRADAS FOR MAIOR QUE O LIMIAR RECEBE 1, SE FOR MENOR RECEE ZERO -> CALCULO E FEITO PARA CADA PESO E ENTRADA INDIVIDALMENTE
            saida =1
        else:
            saida =0

        if(saida !=Y[i]): # SE A SADA FOR DIFERENTE DO VALOR CORRETO, EU CONTINUO O LAO, CALCULANDO OS PESOS
            flag=1
            W = W + n*(Y[i] - saida)*X[i] # ATUALIZANDO O VETOR DE PESOS COM BASE NO ERRO UTILIZANDO A FORMULA DO GRADIENTE APROXIMADO  VARIAO DOS PESOS
    cont = cont+1


plt.scatter(X[Y==0,0],X[Y==0,1])
plt.scatter(X[Y==1,0],X[Y==1,1])

X_1 = np.arange(-0.5,1.5,0.1)
X_2 = -(W[0]/W[1])*X_1 + threshold/W[1]
plt.plot(X_1,X_2)
```

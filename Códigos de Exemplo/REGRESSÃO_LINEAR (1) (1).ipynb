```python
import matplotlib.pyplot as plt
import numpy as np

# Quanto maior os valores de media e desvio padrÃ£o, mais a reta de prediÃ§Ã£o vai ter dificuldade de atingir uma boa precisÃ£o

media = 0
dvp = 1

X = np.array([[1,1],[2,1],[3,1],[4,1]])
Y_novo = Y + np.random.normal(media,dvp,4) # Aplicando um ruÃ­do de media 0 e desvo padrÃ£o 1 na saÃ­da.

# Minimos Quadrados

X_transposta = np.transpose(X)
num= np.dot(X_transposta, Y_novo)
den = np.dot(X_transposta, X)
inv_den = np.linalg.inv(den)
resultado =  np.dot(inv_den,num)

# FunÃ§Ã£o de prediÃ§Ã£o a*x + b*1

x = np.linspace(0,8,100)
y=x*resultado[0] + resultado[1]


plt.scatter(X[:,0],Y,color='yellow') # Printa os pontos (Y,X) apenas a primeira coluna 0 no caso do primeiro
plt.plot(x,y,color='blue') # Printa a funÃ§Ã£o de prediÃ§Ã£o

```

```python
import matplotlib.pyplot as plt
import numpy as np


media = 0
dvp = 0.5

X = np.array([[4,2,1],[16,4,1],[36,6,1],[64,8,1]]) # X**2 + X + 1, mesmo a funÃ§Ã£o sendo X**2 + , Ã© preciso escrever desse modo para que possamos obter os valores dos coeficiente(resultado)
Y = np.array([5,17,37,65])
Y_novo = Y + np.random.normal(media,dvp,4) # Adiconando o ruÃ­do de media 0 e desvio padrÃ£o 0.5

# Minimos Quadrados

X_transposta = np.transpose(X)
num= np.dot(X_transposta, Y_novo)
den = np.dot(X_transposta, X)
inv_den = np.linalg.inv(den)
resultado =  np.dot(inv_den,num)

# FunÃ§Ã£o de prediÃ§Ã£o  AX^2 + BX + C*1 -> X**2 + 1
x = np.linspace(0,8,1000)
y=(x**2)*resultado[0] + resultado[1]*x + resultado[2]


plt.scatter(X[:,1],Y,color='yellow') # Nesse caso nos interessamos apenas na coluna 1 (segunda coluna) pois nela que estÃ¡ o ponto X
plt.plot(x,y,color='blue')


```

```python
import matplotlib.pyplot as plt
import numpy as np


X = np.array([[8,4,2,1],[27,9,3,1],[64,16,4,1],[125,25,5,1]]) # A*X**3 + B*X**2 + C*X + 1*D
Y = np.array([21,50,101,180]) # X**3 + 2*X**2 + 5

# Minimos Quadrados

X_trans = np.transpose(X)
num = np.dot(X_trans,Y)
den = np.dot(X_trans,X)
inv_den = np.linalg.inv(den)

resultado = np.dot(inv_den,num)

# FunÃ§Ã£o de prediÃ§Ã£o

x = np.linspace(0,8,100)
y = resultado[0]*x**3 + resultado[1]*x**2 + resultado[2]*x + resultado[0]*1  # # A*X**3 + B*X**2 + C*X + 1*D -> X**3 + 2*X**2 + 5


# Plots

plt.scatter(X[:,2],Y,color='yellow')
plt.plot(x,y,color='blue')
```

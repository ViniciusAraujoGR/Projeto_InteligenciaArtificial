IRIS dataset COM PCA

```python
#pip install ucimlrepo
import numpy as np
import matplotlib.pyplot as plt

```

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features
y = iris.data.targets


pca_input = X
pca_target = y

#Media dos atributos

media_pca=[]
for column in pca_input.columns:
    media_pca.append(np.mean(pca_input[column]))

central = pca_input - media_pca # Centralização

# Matriz Covariancia

covar=central.cov()

# Autovalores  eAutovetores

eigenvalues, eigenvectors = np.linalg.eig(covar)

# Mudança de dimensão de 4 dimensoes do dataset original para 2 dimensoes

K=2
autovetores_sel= eigenvectors[:,:K]
autovalores_sel= eigenvalues[:K]

nova_projeção = central.dot(autovetores_sel)    # Autovetores x Centralização


color_map = {'Iris-setosa': 'blue', 'Iris-versicolor': 'yellow', 'Iris-virginica': 'green'}
colors = iris.data.targets['class'].map(color_map)

# Plotando o gráfico
plt.scatter(nova_projeção[0],nova_projeção[1], c=colors)
plt.xlabel('Componente 1')
plt.ylabel('Componente')
plt.title('Iris Dataset')
plt.legend()
plt.show()
```

VIDROS DATASET COM PCA

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
glass_identification = fetch_ucirepo(id=42)

# data (as pandas dataframes)
X = glass_identification.data.features
y = glass_identification.data.targets

pca_input = X
pca_target = y

#Media dos atributos

media_pca=[]
for column in pca_input.columns:
    media_pca.append(np.mean(pca_input[column]))

central = pca_input - media_pca # Centralização

# Matriz Covariancia

covar=central.cov()

# Autovalores  eAutovetores

eigenvalues, eigenvectors = np.linalg.eig(covar)

# Mudança de dimensão de 7 dimensoes do dataset original para 3 dimensoes

K=3
autovetores_sel= eigenvectors[:,:K]
autovalores_sel= eigenvalues[:K]

nova_projeção = central.dot(autovetores_sel)    # Autovetores x Centralização

color_map = {1: 'blue', 2: 'red', 3: 'green', 4: 'yellow', 5: 'purple', 6: 'orange', 7: 'brown'}
colors = glass_identification.data.targets['Type_of_glass'].map(color_map)

# Plotando o gráfico
plt.scatter(nova_projeção[0],nova_projeção[1], c=colors)
plt.xlabel('Componente 1')
plt.ylabel('Componente')
plt.title('Iris Dataset')
plt.legend()
plt.show()
```

PCA PLOT 3D

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

# Supondo que você já tenha carregado os dados em um DataFrame chamado "glass_identification"

# Exemplo de mapeamento de cores
color_map = {1: 'blue', 2: 'red', 3: 'green', 4: 'yellow', 5: 'purple', 6: 'orange', 7: 'brown'}

# Supondo que "nova_projeção" seja uma matriz 3D contendo as coordenadas x, y e z para cada ponto

# Criando a figura e o subplot tridimensional
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Mapeando as cores para as classes de vidro
colors = glass_identification.data.targets['Type_of_glass'].map(color_map)

# Plotando o gráfico tridimensional
ax.scatter(nova_projeção[0], nova_projeção[1], nova_projeção[2], c=colors)

# Definindo os rótulos dos eixos
ax.set_xlabel('Componente 1')
ax.set_ylabel('Componente 2')
ax.set_zlabel('Componente 3')

# Adicionando título
plt.title('Glass Classes 3D Scatter Plot')

# Exibindo o gráfico
plt.show()
```

```python
import numpy as np
#X = [1,2,3,4,5]
#ADALINE
# 3X+2 #intercept
#X**1 , X**0 -> 3*X**1 + 2*X**0

X_adj = np.array( [ [1,1], [2,1],[3,1],[4,1],[5,1] ]) # lista com as entradas [[x1,x0]]
noise = np.random.normal(0, 0.2, 5) # ruído para gerar erro
Y = np.array( [5,8,11,14,17] ) + noise  # adiciona ruído na saída
#W = np.random.rand(2) # inicializaçãpo aleatória dos pesos , pode dar um valor bom ou ruim dependendnendo d eond eé iniciado no grafico di n dimensoes do valor dos pesos
W = np.linalg.pinv(X_adj).dot(Y) # inicialização dos pesos por minimos quadrados
n=0.001 # taxa de aprendizagem
erro=0 # inicialização do erro
somatorioErro=0 # inicialização do erro
iteracoes = 1000
for j in range(iteracoes):
  for i in range(len(X_adj)):
    net = X_adj[i,:].dot(W)  # multiplicação do peso*entrada -> função de ativação linear -> gerando o valor que presumis ser o correto

    erro =(Y[i]-net)
    W = W +n*erro*X_adj[i,:]  # Regra do gradiente para calcular o valor dos pesos com base no erro (como se trata de uma função de 0 á 1, sua derivada vai ser sempre 1)
    somatorioErro=somatorioErro+erro**2 # erro
  print('Erro =%f'%(somatorioErro))
  somatorioErro=0
```
